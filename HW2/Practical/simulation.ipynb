```python
x_train, y_train, x_test, y_test = load_data()
print("Fitting the model...")
svm = SVM(eta=0.0001, C=10, niter=200, batch_size=100, verbose=False)
train_losses_10, train_accs_10, test_losses_10, test_accs_10 = svm.fit(x_train, y_train, x_test, y_test)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(train_losses_1, label="C=1")
plt.plot(train_losses_5, label="C=5")
plt.plot(train_losses_10, label="C=10")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(train_accs_1, label="C=1")
plt.plot(train_accs_5, label="C=5")
plt.plot(train_accs_10, label="C=10")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training Accuracy")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(test_losses_1, label="C=1")
plt.plot(test_losses_5, label="C=5")
plt.plot(test_losses_10, label="C=10")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Test Loss")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(test_accs_1, label="C=1")
plt.plot(test_accs_5, label="C=5")
plt.plot(test_accs_10, label="C=10")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Test Accuracy")
plt.legend()
plt.show()
```
