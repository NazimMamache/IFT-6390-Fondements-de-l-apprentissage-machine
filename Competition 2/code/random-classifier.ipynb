```python
train_data = pd.read_csv("../input/kaggle-competition-2/train_data.csv").drop(columns=["id"])
train_labels = pd.read_csv("../input/kaggle-competition-2/train_results.csv").drop(columns=["id"])
test_data = pd.read_csv("../input/kaggle-competition-2/test_data.csv").drop(columns=["id"])
```

```python
def transform_label_to_numeric(label) :
    if label == "negative" :
        return 0
    elif label == "neutral" :
        return 1
    return 2
```

```python
train_labels
```

```python
train_labels["target"] = train_labels["target"].apply(transform_label_to_numeric)
train_labels
```

```python

pos_percentage = len(np.where(train_labels["target"] == 0)[0])/len(train_labels["target"])*100
neg_percentage = len(np.where(train_labels["target"] == 1)[0])/len(train_labels["target"])*100
neut_percentage = len(np.where(train_labels["target"] == 2)[0])/len(train_labels["target"])*100
print(pos_percentage)
print(neg_percentage)
print(neut_percentage)
```

```python
class RandomClassifier :
    
    def predict(self, test_data) :
        return np.random.randint(0, 3, len(test_data))
```

```python
RD = RandomClassifier()
preds = RD.predict(test_data)
preds
```

```python
preds = [[i, preds[i]] for i in range(len(preds))]
df = pd.DataFrame(preds, columns = ["id", "target"])
```

```python
df.to_csv("./predictions_random_3.csv", index = False)
```
